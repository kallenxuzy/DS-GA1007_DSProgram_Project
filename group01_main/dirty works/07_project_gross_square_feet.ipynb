```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading data
bronx = pd.read_excel("/content/drive/My Drive/1007 Project/data/rollingsales_bronx.xls", header = 4)
brooklyn = pd.read_excel("/content/drive/My Drive/1007 Project/data/rollingsales_brooklyn.xls", header = 4)
manhattan = pd.read_excel("/content/drive/My Drive/1007 Project/data/rollingsales_manhattan.xls", header = 4)
queens = pd.read_excel("/content/drive/My Drive/1007 Project/data/rollingsales_queens.xls", header = 4)
staten = pd.read_excel("/content/drive/My Drive/1007 Project/data/rollingsales_statenisland.xls", header = 4)
```

```python
nyc = pd.concat([bronx, brooklyn,manhattan,queens,staten])
```

```python
nyc.columns=[nyc.columns[i].lower().replace(' ','_') for i in range(len(nyc.columns))]
```

Price transformation

```python
import matplotlib.ticker as ticker

plt.figure(figsize=(10,5))
plotd = sns.distplot(nyc[(nyc['sale_price']>100) & (nyc['sale_price'] < 5000000)]['sale_price'], kde=True, bins=100)

tick_spacing=250000 # set spacing for each tick
plotd.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plotd.set_xlim([-100000, 5000000]) # do not show negative values 
plt.xticks(rotation=30) # rotate x ticks by 30 degrees
plt.axvline(nyc[(nyc['sale_price']>100) & (nyc['sale_price'] < 5000000)]['sale_price'].mean(), c='red')
plt.axvline(nyc[(nyc['sale_price']>100) & (nyc['sale_price'] < 5000000)]['sale_price'].median(), c='blue')
plt.text(250000,0.0000012, "median")
plt.text(850000,0.0000010, "mean")
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
nyc = nyc[nyc['sale_price'] != 0]
sns.distplot(nyc['sale_price'])
# Skewness and kurtosis
print("Skewness: %f" % nyc['sale_price'].skew())
print("Kurtosis: %f" % nyc['sale_price'].kurt())
```
